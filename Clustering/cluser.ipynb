# processsing or features

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(np.array(data))
```

```python
X[:3]
```

# Creating our Base model

```python
from sklearn.cluster import KMeans
base_model = KMeans(n_clusters=3)
base_model.fit(X)
```

```python
base_model.predict(X)
```

```python
data.corr()
```

# creating models with different number of clusters and analizing it to reach optimal solution

```python
ks = 20
ssds = []
for i in range(2,ks):
    model = KMeans(n_clusters=i)
    model.fit(X)
    ssds.append(model.inertia_)
```

```python
ssds
```

```python
plt.plot(range(2,20),ssds, '-o')
plt.xlabel('Number of cluster')
plt.ylabel('SSD')
```

```python
plt.plot(range(2, 20), ssds, 'o-')
plt.axhline(504, color='r')
plt.axvline(7, color='g')
plt.xlabel('number of clusters')
plt.ylabel('SSD')
```

```python
ssd_distances = []
for i in range(0, len(ssds)):
    if i == len(ssds) - 1:
        break
    ssd_distances.append(ssds[i+1] - ssds[i])
```

```python
ssd_distances
```

```python
plt.plot(range(2, 19), ssd_distances)
```

```python
best_cluster_number = 6
final_model = KMeans(n_clusters=best_cluster_number)
final_model.fit(X)
```

```python
final_model.predict(X)
```

```python
data['Cluster'] = final_model.predict(X)
data['Country'] = countries
```
