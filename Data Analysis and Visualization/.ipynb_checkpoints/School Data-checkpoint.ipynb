### 3. Visulize all possible charts on given data set.

```python
import seaborn as sns
corr = data.corr()
_, ax = plt.subplots(figsize=(13,10)) 

# graph correlation matrix
_ = sns.heatmap(corr, ax=ax,
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values,
                cmap='coolwarm')
```

### 4. Describe relationship between school_rating and reduced_lunch.

```python
fig, ax = plt.subplots(figsize=(14,8))

ax.set_ylabel('school_rating')

# boxplot with only these two variables
_ = data[['reduced_lunch', 'school_rating']].boxplot(by='school_rating', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
plt.figure(figsize=(14,8)) # set the size of the graph
_ = sns.regplot(data=data, x='reduced_lunch', y='school_rating')
```

### 5. Perform grouping based on school_rating and perform aggregation on reduced_lunch.

```python
data[['reduced_lunch', 'school_rating']].groupby(['school_rating']).describe()
```
